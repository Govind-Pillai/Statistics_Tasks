Standardization


```python
import pandas as pd
import numpy as np
import random 
from scipy import stats
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
actual = pd.DataFrame()
actual["value"] = np.random.randint(10,50,10)
actual["predicted"] = np.random.randint(10,50,10)
```

```python
x_min = actual["value"].min()
mean = actual["value"].mean()
standard = actual["value"].std()
normalization_denominator = (actual["value"].max()) - x_min
count_df = len(actual["value"])
```

```python
def standardization_function(x):
  return (x - mean)/standard
```

```python
actual["standardization"] = actual["value"].apply(standardization_function)
```

Normalization

```python
def normalization_function(x):
  return (x - x_min)/normalization_denominator
```

```python
actual["normalization"] = actual["value"].apply(normalization_function)
```

```python
print(actual.head(10))
```

Variance

```python
def variance(x):
    return ((x - mean)**2)/count_df
```

```python
actual["variance"] = actual["value"].apply(variance)
```

Standard Deviation

```python
def standard_deviation(x):
    return (((x - mean)**2)/count_df)**1/2
```

```python
actual["standard_deviation"] = actual["value"].apply(standard_deviation)
```

```python
print(actual.head(10))
```

Mean Absolute Error

```python
def mean_absolute_error(x):
    return abs(x["value"] - x["predicted"]).mean()
```

```python
actual["mean_absolute_error"] = actual.apply(mean_absolute_error,axis = 1)
```

```python
print(actual.head(10))
```

Random Sampling

```python
salary = pd.DataFrame(np.random.randint(20000,80000,100))
```

```python
def averageOfgroups(df,size):
    average_of_groups = []
    for i in range(0,size):
        average_of_groups.append(np.mean(df[i]))
    return np.mean(average_of_groups)
```

Without replacement

```python
def without_replacement(data):
    groups = []
    for i in range(0, len(data), 15):
        groups.append(data[i:i + 15])
    return groups
```

```python
group_without_replacement = without_replacement(salary[0])
```

```python
print(averageOfgroups(group_without_replacement,7))
```

```python
salary[0].sample(8)
```

With Replacement

```python
def with_replacement(data):
    groups = []
    for i in range(15):
        row = []
        for j in range(30):
            row.append(data[np.random.randint(0, len(data))])
        groups.append(row)
    return groups
```

```python
group_with_replacement = with_replacement(salary[0])
```

```python
print(averageOfgroups(group_with_replacement,15))
```

left tailed test
A factory produces wires with an average tensile strength of 500 N. A new supplier sends cheaper raw material. The engineer wants to check if the tensile strength has decreased.

H₀: μ = 500

Hₐ: μ < 500 (left-tailed test)

```python
battery_df=pd.DataFrame(np.random.randint(0,10,30))
```

```python
n=30
population_mean=30
std_population=3
alpha=0.05
sample_mean=battery_df[0].mean()
```

```python
z_sample = (sample_mean - population_mean) / (std_population / (n**0.5))
z_critical = stats.norm.ppf(alpha)
if z_sample < z_critical:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

A beverage company claims their soda cans contain 330 ml on average. A quality analyst wants to check if the filling machine deviates (could be more or less).

H₀: μ = 330

Hₐ: μ ≠ 330 (two-tailed test)

```python
biscuit_df=pd.DataFrame(np.random.randint(0,10,30))
n=30
population_mean=30
std_population=3
alpha=0.05
sample_mean=biscuit_df[0].mean()
```

```python
z_sample = (sample_mean - population_mean) / (std_population / (n**0.5))
z_critical_right = stats.norm.ppf(1 - alpha/2)
z_critical_left = stats.norm.ppf(alpha/2)
if z_critical_left > z_sample > z_critical_right:
        print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
battery_life = pd.DataFrame(np.random.randint(0,10,30))

# Given
n = len(battery_life)
sample_mean = battery_life[0].mean()
sample_std = battery_life[0].std()
population_mean = 7
alpha = 0.05

# t statistic
t_data = (sample_mean - population_mean) / (sample_std / (n**0.5))

# t critical (left-tailed)
t_critical_left = stats.t.ppf(alpha, df=n-1)

# comparison
if t_data < t_critical_left:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
dominos_delivery = pd.DataFrame(np.random.randint(15,45,30))
pizza_delivery=pd.DataFrame(np.random.randint(15,45,35))
#Given
sample_dominos=30
sample_pizza=35
sample_dominos_mean=dominos_delivery[0].mean()
sample_pizza_mean=pizza_delivery[0].mean()
sample_dominos_std=dominos_delivery[0].std()
sample_pizza_std=pizza_delivery[0].std()

#s_data calculate
s_data=((sample_dominos-1)*(sample_dominos_std**2)+(sample_pizza-1)*(sample_pizza_std**2))/(sample_pizza+sample_dominos-2)

#t_data
t_data=(sample_dominos_mean-sample_pizza_mean)/(s_data*((1/sample_dominos)+(1/sample_pizza))**0.5)

#t_critical
t_critical_right=stats.t.ppf(1-alpha/2,df=sample_dominos+sample_pizza-2)
t_critical_left=stats.t.ppf(alpha/2,df=sample_dominos+sample_pizza-2)

#comparision
if t_data>t_critical_right or t_data<t_critical_left:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
expected = pd.Series(np.random.randint(5,10,30))
actual   = pd.Series(np.random.randint(5,10,30))

# difference
diff = actual - expected

n = len(diff)
diff_mean = diff.mean()
diff_std = diff.std()

alpha = 0.05

# t statistic
t_data = diff_mean / (diff_std / (n**0.5))

# t critical two-tailed
t_critical_right = stats.t.ppf(1 - alpha/2, df=n-1)
t_critical_left  = stats.t.ppf(alpha/2, df=n-1)

# decision
if t_data > t_critical_right or t_data < t_critical_left:
    print("Reject H0 (difference exists)")
else:
    print("Fail to reject H0 (no difference)")
```
